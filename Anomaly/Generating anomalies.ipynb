```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
signal = np.random.normal(scale=3, size=200)
signal[:50] += 50; signal[50:] += 60; signal[150:] -= 20
signal[110] = 0
signal[75] = 80

plt.plot(signal)
plt.ylim(0,100)
plt.show()
```

```python
signal = np.sin(np.linspace(0, 40*np.pi, num=300))
signal[105:155] *= 0.1
signal[250:280] *= 0.5
signal = 12 * signal + 50

noise = np.random.normal(scale = 5, size=300)
signal = signal + noise

plt.ylim(0,100)
plt.plot(signal)
plt.show()
```

```python
signal = np.sin(np.linspace(0, 50*np.pi, num=300))
signal[105:155] *= 0.1
signal[250:280] *= 0.3
signal[55] = 4
signal[10:50] = signal[10]
signal = 12 * signal + 50

noise = np.random.normal(scale = 5, size=300)
signal = signal + noise

plt.ylim(0,100)
plt.plot(signal)
plt.show()
```

```python
x = np.linspace(0, 1000,1000)
y = np.sin(0.05*x) + 0.3*np.sin(0.3*x)
y[200] = 3
y[300:320] = 0
y[500:550] *= 0.3
y[800:830] *= 0.1
y[650:655] = 2
y[80:100] = -1
plt.figure(figsize=(10,10))
plt.plot(x,y)
```
