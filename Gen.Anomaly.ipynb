```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
signal = np.random.normal(scale=3, size=200)
signal[:50] += 50; signal[50:] += 60; signal[150:] -= 20
signal[110] = 0
signal[75] = 80

plt.plot(signal)
plt.ylim(0,100)
plt.show()
```

```python
signal = np.sin(np.linspace(0, 40*np.pi, num=300))
signal[105:155] *= 0.1
signal[250:280] *= 0.5
signal = 12 * signal + 50

noise = np.random.normal(scale = 5, size=300)
signal = signal + noise

plt.ylim(0,100)
plt.plot(signal)
plt.show()
```

```python
signal = np.sin(np.linspace(0, 50*np.pi, num=300))
signal[105:155] *= 0.1
signal[250:280] *= 0.3
signal[55] = 4
signal[10:50] = signal[10]
signal = 12 * signal + 50

noise = np.random.normal(scale = 5, size=300)
signal = signal + noise

plt.ylim(0,100)
plt.plot(signal)
plt.show()
```

```python
signal = np.sin(np.linspace(0, 40*np.pi, num=300))
signal2 = np.sin(np.linspace(0, 30*np.pi, num=300))
signal[105:155] *= 0.1
signal[250:280] *= 0.1
signal[55] = 4
signal[10:30] = signal[10]
signal = 12 * signal + 50


signal2[250:280] *= 0.3

signal2[10:30] = signal2[10]
signal2 = 12 * signal2 + 50

noise = np.random.normal(scale = 3, size=300)
signal = signal + noise
signal2 = signal2 + noise

plt.ylim(0,100)
plt.plot(signal, color="blue")
plt.plot(signal2, color="blue")
plt.show()
```
